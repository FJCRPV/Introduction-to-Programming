Data are in constant 2010 U.S. dollars because development encompasses many factors - economic, environmental, cultural, educational, and institutional - no single measure gives a complete picture. However well-being might be different for each level of income, later on you will be asked for a summary table and a graph to report the level of income by each 20% of the population. 

[The World Bank](https://databank.worldbank.org/source/world-development-indicators)  constructs the **Gini index**, which measures the extent to which the distribution of income (or, in some cases, consumption expenditure) among individuals or households within an economy deviates from a perfectly equal distribution.

<div class="alert alert-block alert-info">
<b>Question 10:</b> Graph the scatter plot between <code>GNI (constant 2010 US$)_mean</code> and  <code>GINI index (World Bank estimate)_mean</code>:

scatter

```python
# your code here
x = df_8['GINI index (World Bank estimate)_mean'].to_numpy()
y = df_8['GNI (constant 2010 US$)_mean'].to_numpy()

plt.scatter(x, y, color='orange')

plt.xlabel('GINI Index')
plt.ylabel('GNI $')
plt.title('GNI Vs.GINI')
plt.xlim(24,56)

plt.show()
```

<br><br>

A Lorenz curve plots the cumulative percentages of total income received against the cumulative number of recipients, starting with the poorest individual or household. The Gini index measures the area between the Lorenz curve and a hypothetical line of absolute equality, expressed as a percentage of the maximum area under the line. 

**Thus a Gini index of 0 represents perfect equality, while an index of 100 implies perfect inequality.**

```python
# your code here
def one_hot_encode(df,column_name):
    new_df = pd.get_dummies(df[column_name],prefix=column_name) 
   
    df = pd.concat([df,new_df],axis=1)

    df.drop([column_name], axis=1, inplace=True)

    return df

df = one_hot_encode(df,'Continent')
```

In machine learning we are always looking for an algorithm to enter into production stage, meaning receiving new data points and being able to predict the target variable with the coeficients already trained. That's why we care a lot about a concept called *data leakage*.

Data leakage (or leakage) happens when your training data contains information about the target, but similar data will not be available when the model is used for prediction. This leads to high performance on the training set (and possibly even the validation data), but the model will perform poorly in production. For a complete description I recomend the reader to see this free [Kaggle Course](https://www.kaggle.com/learn/intermediate-machine-learning).

<b>Question 25:</b> Define a function to measure the quality of each approach, named it <code>score_dataset(X_train, X_test, y_test, y_test, model="linear")</code>. 
</div>



`score_dataset(X_train, X_test, y_test, y_test, model="linear")` should allow to compare different approaches to dealing with missing values. This function should reports the [mean absolute error] and have as a parameter the type of model to be created from the following list:

        model = LinearRegression() 
        model = DecisionTreeRegressor(random_state=0)
        model = RandomForestRegressor(random_state=0)
        
The default value of the parameter must be "linear_model". The possible values for it should be: "linear", "decision_tree","random_forest".

<br><br>

```python
# your code here
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

def score_dataset(X_train,X_test,y_train,y_test, model='linear'):
    if model=='linear':
        model = LinearRegression()
    elif model=='decision_tree':
        model = DecisionTreeRegressor(random_state=0)
    elif model=='random_forest':
        model = RandomForestRegressor(random_state=0)
        
    model.fit(X_train,y_train) 
    preds=model.predict(X_test)
    return mean_absolute_error(y_test,preds)
```
